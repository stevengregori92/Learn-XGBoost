```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value
from jcopml.feature_importance import mean_score_decrease
```

```python
df = pd.read_csv('dairy_dataset.csv', index_col = 'Product ID')
pd.options.display.max_columns = 50
df
```

#Cleaning Dataset

```python
plot_missing_value(df, return_df=True)
```

```python
df.drop(columns=['Date', 'Production Date', 'Expiration Date'], inplace=True)
```

```python
df.info()
```

#Dataset Splitting

```python
X = df.drop(columns=['Price per Unit (sold)'])
y = df['Price per Unit (sold)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#Training Section

```python
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
```

```python
X_train.columns
```

```python
df.iloc[0:2]
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(), ['Total Land Area (acres)', 'Number of Cows', 'Quantity (liters/kg)',
                             'Price per Unit', 'Total Value', 'Shelf Life (days)',
                             'Quantity Sold (liters/kg)', 'Approx. Total Revenue(INR)',
                             'Quantity in Stock (liters/kg)', 'Reorder Quantity (liters/kg)',
                             'Minimum Stock Threshold (liters/kg)']),
    ('categoric', cat_pipe(), ['Location', 'Farm Size', 'Product Name', 'Brand', 'Customer Location',
                               'Sales Channel'])
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', XGBRegressor(n_jobs=-1, random_state=42))
])

model = RandomizedSearchCV(pipeline, rsp.xgb_params, cv=3, n_iter=50, n_jobs=-1, verbose=1, random_state=42)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

#Feature Importance

```python
mean_score_decrease(X_train, y_train, model, plot=True)
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(), ['Quantity (liters/kg)', 'Price per Unit', 'Total Value',
                             'Quantity Sold (liters/kg)', 'Approx. Total Revenue(INR)'])
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', XGBRegressor(n_jobs=-1, random_state=42))
])

model = RandomizedSearchCV(pipeline, rsp.xgb_params, cv=3, n_iter=50, n_jobs=-1, verbose=1, random_state=42)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```
